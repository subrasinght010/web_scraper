```python
from bs4 import BeautifulSoup as bf
import requests
import pandas as pd
import csv
```

```python
def scrape_pages(num_pages):
    all_product_data = []
    for page_num in range(1, num_pages + 1):
        product_data = scrape_page(page_num)
        for i in product_data:
            all_product_data.append(i)
    return all_product_data

def scrape_page(page_num):
    headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36"
     }
    URL = f"https://www.amazon.in/s?k=bags&page={page_num}&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_{page_num}"
    response = requests.get(URL , headers=headers).text
    soup  = bf(response,'lxml')
    print(page_num)
    return extract_info(soup)


def scrape_page1( URL):
    headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36"
     }
    response = requests.get(URL , headers=headers).text
    soup  = bf(response,'lxml')
    return  extract_product_details(soup)
        
    
def write_to_csv(data, filename):
    with open(filename, "w", newline="", encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=data[0].keys())
        writer.writeheader()
        for item in data:
            writer.writerow(item)  
            

            

def extract_info(soup):
    
    product_data = []
    products = soup.find_all("div", {"data-index": True})
    for product in products:

        product_url = product.find('a', {'class': 'a-link-normal'})
        if product_url:
            #print(product_url['href'].strip())
            product_url=is_valid_url(product_url['href'].strip())
            #print(product_url)
            description,product_description ,product_asin,product_manufacturer = scrape_page1(product_url)

        else:
            product_url = 'Not available'
            description = 'Not available'
            product_description ='Not available'
            product_asin = 'Not available'
            product_manufacturer = 'Not available'

        product_name = product.find('span', {'class': 'a-size-medium a-color-base a-text-normal'})
        if product_name:
            product_name = product_name.text.strip()
        else:
            product_name = 'Not available'

        product_price = product.find('span', {'class': 'a-offscreen'})
        if product_price:
            product_price = product_price.text.strip()
        else:
            product_price = 'Not available'

        product_rating = product.find('span', {'class': 'a-icon-alt'})
        if product_rating:
            product_rating = product_rating.text.strip()
        else:
            product_rating = 'Not available'

        product_reviews = product.find('span', {'class': 'a-size-base'})
        if product_reviews:
            product_reviews = product_reviews.text.strip()
        else:
            product_reviews = 'Not available'
        
        
        
        
        product_data.append({
                "product_url": product_url,
                "product_name": product_name,
                "product_price": product_price,
                "product_rating": product_rating,
                "product_reviews": product_reviews,
                "product_description": product_description,
                "product_asin": product_asin,
                "product_manufacturer":product_manufacturer,
                "description": description,

            })
    return product_data


def extract_product_details(soup ):
       
    description = soup.find("div", id="feature-bullets")
    if description:
        description = description.text.strip()
    else:
        description = 'Not available'
    
    product_description = soup.find("div", id="productDescription")
    if product_description:
        product_description = product_description.text.strip()
    else:
        product_description = 'Not available'
            
    product_asin = soup.find("input", id="ASIN")       
    if product_asin:
        product_asin= product_asin["value"] 
    else:
        product_asin = 'Not available'
    
    
    manufacturer = soup.find("a", {"class": "a-link-normal"})
    if manufacturer:
        manufacturer =manufacturer.text.strip()
    else:
        manufacturer = 'Not available'
    
    return description,product_description ,product_asin,manufacturer

def is_valid_url(url):
    if "https://www.amazon.in/"  in url or "http://www.amazon.in/" in url :
        return url
    else:
        return "https://www.amazon.in"+url
    

```
